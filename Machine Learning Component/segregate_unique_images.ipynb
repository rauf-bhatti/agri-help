```python
import cv2
import numpy as np
import os
import shutil
```

```python
def move_photos(PATH_TO_DATA, file_name):
    new_path = os.path.join(PATH_TO_DATA, "Unique_Dataset")
    isExist = os.path.exists(new_path)
    
    
    if not isExist:
        os.makedirs(new_path)
        print("NEW DIR CREATED!")
    else:
        print("NEW DIR PRESENT!")
    
    shutil.move(file_name, new_path)
        
```

```python
def load_training_data():
    # Query the data in the OS
    PATH_TO_DATA = "Cotton_Leaf_Dataset/Pakistan_Dataset"
    # Alternaria
    DISEASE_CLASSES = ["Bacterial_Disease"] #, "Fungal_Disease", "Viral_Disease", "Healthy"]#, "Bacterial Blight", "Cercospora"]
    
    training_data = []
    
    for disease_class in DISEASE_CLASSES:
        path_to_class = os.path.join(PATH_TO_DATA, disease_class)
        path_to_class_unique = os.path.join(path_to_class, "Unique_Dataser")
        img_array_rgb_prev = []
        index = 0
        same_images = 0
        different_images = 0
        
        for img in os.listdir(path_to_class):
            if img == ".DS_Store":
                continue
            # Read image using cv2
            
            picture_path = os.path.join(path_to_class, img)
            img_array_rgb = cv2.imread(picture_path)
            #normalize the image
            img_array_rgb_resized = cv2.resize(img_array_rgb, (120, 120))
            
            if index != 0:
                diff = cv2.subtract(img_array_rgb_resized, img_array_rgb_prev)
                result = not np.any(diff)
                if result is True:
                    print("Pictures are the same")
                    same_images = same_images + 1
                else:
                    different_images = different_images + 1
                    print("Pictures are different")
                    move_photos(path_to_class, picture_path)
            
            img_array_rgb_prev = img_array_rgb_resized
            
            #plt.imshow(img_array_rgb_resized)
            #plt.show()
            
            # img_array_gs = cv2.imread(os.path.join(path_to_class, img), cv2.IMREAD_GRAYSCALE)
            #index = DISEASE_CLASSES.index(disease_class)
            
            # Correlate the image RGB array and the classification
            #training_data.append([img_array_rgb_resized, index])
            
            index = index + 1
            
        print ("Total same images: ", same_images) 
        print ("Different images: ", different_images)
    return training_data 
```
